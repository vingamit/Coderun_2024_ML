```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("data/easy_pursuit_of_profit_input.txt", 'r') as file:
    k, n, c_cost, c = map(int, file.readline().rstrip().split())
    xy = [list(map(int, file.readline().rstrip().split())) for _ in range(n)]
```

```python
pairs = np.array(xy)
```

```python
plt.scatter(pairs[:, 0], pairs[:, 1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=k)
```

```python
cluster_labels = kmeans.fit_predict(xy)
```

```python
plt.scatter(xy[:, 0], xy[:, 1], c=cluster_labels, cmap='virdis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Data with Clusters')
plt.colorbar(label='Cluster')
plt.show()
```
